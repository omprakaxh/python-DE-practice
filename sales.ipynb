```python
import pandas as pd
```

```python
df = pd.read_csv('sales_activity.csv')
df
```

```python
df['Revenue'] = df['Quantity'] * df['Price']
df
```

```python
df.groupby('Region')['Revenue'].agg('sum')
```

```python
df.groupby('Product')['Revenue'].agg("sum")
```

```python
df
```

```python
df.groupby('Month')['Revenue'].agg("sum")
```

```python
month_order = ["January", "February", "March", "April", "May"]

result = df.groupby("Month")["Revenue"].sum()
result = result.reindex(month_order)  # ✅ reorder by correct month sequence
print(result)
```

```python
df
```

```python
# Step 1 - Total revenue per region
total_revenue = df.groupby("Region")["Revenue"].sum()

# Step 2 - Best product per region (product with highest revenue)
best_product = df.loc[df.groupby("Region")["Revenue"].idxmax()].set_index("Region")["Product"]

# Step 3 - Combine into a new DataFrame
summary = pd.DataFrame({
    "Total Revenue": total_revenue,
    "Best Product": best_product
}).reset_index()

print(summary)

# Step 4 - Save to CSV
summary.to_csv("sales_summary.csv", index=False)
print("✅ sales_summary.csv saved!")
```
